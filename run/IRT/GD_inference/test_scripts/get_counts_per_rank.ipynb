```python
import pandas as pd

df = pd.read_csv("/root/jw/TAPTAP_IRT/run/IRT/GD_inference/integration_test/compiled_data_with_penalty_2023.csv")
# Sample data as a string
df
```

```python
# Group by 'Subject' and 'Rank' and calculate counts and percentages
rank_counts = df.groupby(['Subject', 'Rank']).size().reset_index(name='Count')

# Calculate percentages within each Subject
rank_counts['Percentage'] = rank_counts.groupby('Subject')['Count'].transform(lambda x: (x / x.sum()) * 100)
```

```python

# Display the results
result = rank_counts.pivot(index='Subject', columns='Rank', values=['Count', 'Percentage'])
result = result.fillna(0)  # Fill NaN with 0 for subjects with missing Ranks
result
```
